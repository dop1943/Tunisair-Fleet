```python
#ce script python a été utilisé pour scraper des données du site airTrax qui est une pointure mondiale dans la récolte des
#avis d'utlisateurs sur les compagnies aeriennes

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import csv
from math import ceil


csv_file = open('testcloud.txt', 'w',encoding='utf8')
review_dict={
    "customer_review":"1"
    }

writer = csv.DictWriter(csv_file,review_dict.keys(),delimiter="~")
writer.writeheader()
chromedriver = "C:\chromedriver.exe" 
driver = webdriver.Chrome(chromedriver)
 
US_airlines = ["tunisair/","nouvelair/", "air-algerie/", "delta-air-lines/", "royal-air-maroc/", "egyptair/", "royal-jordanian-airlines/", "royal-jordanian-airlines", "emirates/", "qatar-airways/", "jetairfly/", "alitalia/","lufthansa/", "condor-airlines/","brussels-airlines/", "condor-airlines/", "air-malta/","turkish-airlines/","germanwings/"]

US_airline_pages = ["http://www.airlinequality.com/airline-reviews/" + airline for airline in US_airlines]

for page in US_airline_pages:
	driver.get(page)
	time.sleep(0.2)		
	try:
		print("="*40)
		print("Scraping " + page)
			

		review_count = driver.find_element_by_xpath('//div[@class = "rating-totals"]//span[@itemprop = "reviewCount"]').text
		review_count = float(review_count)
		n = min(10,int(ceil(review_count/10)))
	
		index = 1
		while index <= n:
			driver.get(page + "page/" + str(index) +'/')
			time.sleep(5)
			
			try:
				print("Scraping Page number " + str(index))
				index = index + 1
		
				reviews = driver.find_elements_by_xpath('//article[@itemprop = "review"]')
				for review in reviews:
				

					try:	
						customer_review = review.find_element_by_xpath('.//div[@itemprop = "reviewBody"]').text
					except: 
						customer_review = ""

					review_dict['customer_review'] = customer_review
					writer.writerow(review_dict)
			

			except Exception as e:
				print(e)
				csv_file.close()
				break
				

	except Exception as e:
				print(e)
				csv_file.close()
				#driver.close()
				break

csv_file.close()

driver.close()
```
